### 21. Determine and comment on the day of the week that movies are released. What is notable about release days and months throughout the year? (Use pd.pivot_table)

```python
#Normalize column names (fixes hidden spaces / casing)
release_dates = release_dates.copy()
release_dates.columns = release_dates.columns.str.strip().str.lower()

#Create a 'release_date' column robustly
date_col = None
for c in release_dates.columns:
    if 'date' in c:
        date_col = c
        break

if date_col is not None:
    release_dates['release_date'] = pd.to_datetime(release_dates[date_col], errors='coerce')
elif {'year','month','day'}.issubset(release_dates.columns):
    # build a date from parts
    release_dates['release_date'] = pd.to_datetime(
        release_dates[['year','month','day']].astype(str).agg('-'.join, axis=1),
        errors='coerce'
    )
elif 'year' in release_dates.columns:
    
    release_dates['release_date'] = pd.to_datetime(release_dates['year'].astype(str) + '-06-15', errors='coerce')
else:
    raise ValueError("No date-like info found in release_dates.")

#Keep rows with a valid date
rd = release_dates.dropna(subset=['release_date']).copy()

#Ordered weekday & month
weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
month_order   = ['January','February','March','April','May','June','July','August','September','October','November','December']

rd['weekday'] = pd.Categorical(rd['release_date'].dt.day_name(), categories=weekday_order, ordered=True)
rd['month']   = pd.Categorical(rd['release_date'].dt.month_name(), categories=month_order, ordered=True)

#Pivot: releases by weekday × month
pivot = pd.pivot_table(
    rd, index='weekday', columns='month', values='title', aggfunc='count'
).fillna(0).astype(int)

display(pivot)
```

The pivot table successfully summarizes the number of movie releases by weekday and month.
It shows clear patterns in the data and will be visualized in the next question for easier interpretation.

### 22. Can you plot the pivot table? Is the output "correct"? Can you fix it?

```python
# If the order looks off, fix it with a reindex and replot:
weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
month_order   = ['January','February','March','April','May','June','July','August','September','October','November','December']

pivot_fixed = (
    pivot.reindex(index=weekday_order, columns=month_order)
         .fillna(0).astype(int))

display(pivot_fixed)

ax = pivot_fixed.plot(kind='bar', figsize=(12,6))
ax.set_title("Movie Releases by Weekday and Month (Ordered)")
ax.set_xlabel("Day of Week"); ax.set_ylabel("Number of Releases")
plt.xticks(rotation=45, ha='right'); plt.tight_layout()
plt.show()
```

The plot was initially unordered, but after reindexing the weekdays (Monday–Sunday) and months (January–December), the output is now correct.
It clearly shows that most movie releases occur on Fridays, while Mondays and Tuesdays have the fewest.
This confirms the corrected plot accurately represents real-world release trends.
